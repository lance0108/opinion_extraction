```python
import pandas as pd
import gensim
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
import nltk
import pyLDAvis.gensim
import warnings
import jieba
warnings.filterwarnings('ignore')
```

```python
df_data = pd.read_csv(r"C:\Dropbox\_projects\PG\ds-nlp-interview-question_v2.csv")
print(len(df_data))
df_data.head()
```

## English

```python
# Tokenize
sw = set(stopwords.words("english"))
tokenizer = RegexpTokenizer("[a-z]+")
def tokenize(text):
    tokens = tokenizer.tokenize(text.lower())
    return [token for token in tokens if token not in sw]
docs = df_data["REVIEW_TEXT_EN"].apply(tokenize)
```

```python
docs[:5]
```

```python
id2word = gensim.corpora.Dictionary(docs)
id2word.filter_extremes(no_below=3, no_above=0.2)
print(id2word)
```

```python
corpus = [id2word.doc2bow(doc) for doc in docs]
```

```python
n_topics = 10
lda = gensim.models.ldamodel.LdaModel(corpus, num_topics=n_topics, id2word=id2word, passes=10)
```

```python
lda_vis = pyLDAvis.gensim.prepare(lda, corpus, id2word, sort_topics=False)
pyLDAvis.display(lda_vis)
```

## Chinese

```python
# Tokenize
docs = df_data["REVIEW_TEXT_CN"].apply(lambda x: list(jieba.cut(x)))
docs[:5]
```

```python
id2word = gensim.corpora.Dictionary(docs)
id2word.filter_extremes(no_below=3, no_above=0.1)
print(id2word)
```

```python
corpus = [id2word.doc2bow(doc) for doc in docs]
```

```python
n_topics = 20
lda = gensim.models.ldamodel.LdaModel(corpus, num_topics=n_topics, id2word=id2word, passes=10)
```

```python
lda_vis = pyLDAvis.gensim.prepare(lda, corpus, id2word, sort_topics=False)
pyLDAvis.display(lda_vis)
```
